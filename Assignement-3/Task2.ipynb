```python
import pandas as pd
from scipy.stats import pearsonr
import matplotlib.pyplot as plt
```

p2a file 


```python
p2a_df = pd.read_csv("p2a.csv", header = None)
x = p2a_df[0]
y = p2a_df[1]
```

```python
r, p = pearsonr(x,y)
print (f"Pearson Correlation: {r}")
print (f"p-value: {p}")
```

p2b file 


```python
p2b_df = pd.read_csv("p2b.csv", header = None)
a = p2b_df[0]
b = p2b_df[1]
```

```python
r1, p1 = pearsonr(a,b)
print (f"Pearson Correlation: {r1}")
print (f"p-value: {p1}")
```

p2a file scatter plot

```python
plt.scatter (x,y, alpha = 0.5)
plt.xlabel('Variable 1')
plt.ylabel('Variable 2')
plt.title('Scatterplot of Variables in p2a')
plt.savefig('ScatterPlot_p2a.png')
plt.show()
```

p2b file scatter plot

```python
plt.scatter (a,b, alpha = 0.5)
plt.xlabel('Variable 1')
plt.ylabel('Variable 2')
plt.title('Scatterplot of Variables in p2b')
plt.show()
plt.savefig('ScatterPlot_p2b.png')
```
